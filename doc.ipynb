```python
import matplotlib.pyplot as plt

# Metrics for the first model
model1_metrics = {
    "Rouge Recall": 0.5321807821807822,
    "Rouge Precision": 0.5378713599372487,
    "Rouge F1": 0.5181034434385339,
    "Bleu": 0.0078016393687140835,
    "Meteor": 0.47597500000000004,
}

# Metrics for the second model
model2_metrics = {
    "Rouge Recall": 0.7097578347578347,
    "Rouge Precision": 0.7333887043189369,
    "Rouge F1": 0.7198564543421889,
    "Bleu": 0.010165635448451789,
    "Meteor": 0.771675,
}

# Extract metric names and values
metrics = list(model1_metrics.keys())
model1_values = list(model1_metrics.values())
model2_values = list(model2_metrics.values())

# Create bar plots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Bar plot for Rouge Recall
axs[0, 0].bar(
    ["Model 1", "Model 2"],
    [model1_values[0], model2_values[0]],
    color=["blue", "red"],
)
axs[0, 0].set_title("Rouge Recall")

# Bar plot for Rouge Precision
axs[0, 1].bar(
    ["Model 1", "Model 2"],
    [model1_values[1], model2_values[1]],
    color=["blue", "red"],
)
axs[0, 1].set_title("Rouge Precision")

# Bar plot for Rouge F1
axs[0, 2].bar(
    ["Model 1", "Model 2"],
    [model1_values[2], model2_values[2]],
    color=["blue", "red"],
)
axs[0, 2].set_title("Rouge F1")

# Bar plot for Bleu Score
axs[1, 0].bar(
    ["Model 1", "Model 2"],
    [model1_values[3], model2_values[3]],
    color=["blue", "red"],
)
axs[1, 0].set_title("Bleu Score")

# Bar plot for Meteor Score
axs[1, 1].bar(
    ["Model 1", "Model 2"],
    [model1_values[4], model2_values[4]],
    color=["blue", "red"],
)
axs[1, 1].set_title("Meteor Score")

# Remove empty subplot
fig.delaxes(axs[1, 2])

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
